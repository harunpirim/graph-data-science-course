```python
import pandas as pd
import networkx as nx

def filter_empty_attributes(row_dict):
    return {k: v for k, v in row_dict.items() if pd.notnull(v)}

# Load nodes and edges data
nodes_df = pd.read_csv('/Users/harunpirim/Gephi/nodes_msup2.csv')
edges_df = pd.read_csv('/Users/harunpirim/Gephi/edges_msup2.csv')

# Create an empty graph
G = nx.Graph()

# Add nodes to the graph
for index, row in nodes_df.iterrows():
    G.add_node(row['Id'], **filter_empty_attributes(row.to_dict()))

# Add edges to the graph
for index, row in edges_df.iterrows():
    G.add_edge(row['Source'], row['Target'], **filter_empty_attributes(row.to_dict()))

# Now you should be able to write to GraphML
nx.write_graphml(G, '/Users/harunpirim/Gephi/msup2.graphml')
nx.write_gexf(G, '/Users/harunpirim/Gephi/msup2.gexf')
```
